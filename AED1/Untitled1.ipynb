```python
import random

s_number = random.randint(1, 50)

while True:
    palpite = int(input("Digite um valor: "))

    if palpite == s_number:
        print("aeee, ótimo! 🎉")
        break
    elif palpite > s_number:
        print("tente um número menor 🔽")
    else:
        print("tente um número maior 🔼")
```
